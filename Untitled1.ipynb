```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.layers import *
from keras.models import *
from keras.preprocessing import image
from tensorflow.keras import models,layers
import os
import glob
import cv2
```

```python
print(os.listdir("Dataset_150"))
print(glob.glob('Dataset_2000/train/*'))
print(glob.glob('Dataset_2000/valid/*'))
classes = ["NoCancer","Cancer"]
```

```python
train_images=[]
train_labels=[]
for directory_path in glob.glob('Dataset_2000/train/*'):
  label=directory_path.split("//")[-1]
  print(label)
  for img_path in glob.glob(os.path.join(directory_path,"*.png")):
    print(img_path)
    img=cv2.imread(img_path,cv2.IMREAD_COLOR)
    img=cv2.resize(img,(300,300))
    img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    train_images.append(img)
    train_labels.append(label)
```

```python
test_images=[]
test_labels=[]
for directory_path in glob.glob('Dataset_2000/valid/*'):
  label=directory_path.split("//")[-1]
  print(label)
  for img_path in glob.glob(os.path.join(directory_path,"*.png")):
    print(img_path)
    img=cv2.imread(img_path,cv2.IMREAD_COLOR)
    img=cv2.resize(img,(300,300))
    img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    test_images.append(img)
    test_labels.append(label)
```

```python
train_images=np.asarray(train_images)
train_labels=np.asarray(train_labels)
test_images=np.asarray(test_images)
test_labels=np.asarray(test_labels)
train_images.shape
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
le.fit(test_labels)
test_labels_encoded=le.transform(test_labels)
le.fit(train_labels)
train_labels_encoded=le.transform(train_labels)
```

```python
x_train,y_train,x_test,y_test=train_images,train_labels_encoded,test_images,test_labels_encoded
```

```python
X_train = x_train / 255.0
X_test = x_test / 255.0
X_train.shape
```

```python
from tensorflow.keras.applications import EfficientNetB3

NUM_CLASSES = 2
IMG_SIZE = 300
size = (IMG_SIZE, IMG_SIZE)

inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))

#Using model without transfer learning

outputs = EfficientNetB3(include_top = True, weights = None, classes=NUM_CLASSES)(inputs)
```

```python
model = tf.keras.Model(inputs, outputs)
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
r=model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test,y_test)
```
